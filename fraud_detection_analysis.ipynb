```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("C:\\Users\\Bamawo Richard\\Desktop\\PYTHON\\__pycache__\\FRAUDTEST\\fraudTest.csv")

# Convert date columns
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
df["dob"] = pd.to_datetime(df["dob"])

# Create transaction_hour
df["transaction_hour"] = df["trans_date_trans_time"].dt.hour

# Create customer_age
df["customer_age"] = (
    df["trans_date_trans_time"].dt.year - df["dob"].dt.year
)

# Create amount_band
df["amount_band"] = pd.cut(
    df["amt"],
    bins=[0, 100, 200, 300, 500, float("inf")],
    labels=["0-99", "100-199", "200-299", "300-499", "500+"],
    right=False
)

# Haversine formula for distance_to_merchant
def haversine(lat1, lon1, lat2, lon2):
    r = 6371  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    return 2 * r * np.arcsin(np.sqrt(a))

df["distance_to_merchant"] = haversine(
    df["lat"], df["long"], df["merch_lat"], df["merch_long"]
)

# Save prepared file
df.to_csv("fraudTest_prepared.csv", index=False)

print("Done. New file saved as fraudTest_prepared.csv")
print("Rows:", len(df))
print("Columns:", df.columns.tolist())
```
